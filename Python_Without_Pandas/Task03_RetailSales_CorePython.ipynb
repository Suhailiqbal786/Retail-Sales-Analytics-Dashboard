# Count Missing Values

```python
import csv

# Dictionary to store missing value counts
missing = {
    "CustomerName": 0,
    "Email": 0,
    "Phone": 0,
    "City": 0,
    "Rating": 0
}

# Read the CSV file
with open("Orders_Cleaned.csv", mode="r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        for column in missing:
            value = row[column].strip()

            if value == "" or value == "Unknown":
                missing[column] += 1

# Display results
print("Missing Values:")
for column, count in missing.items():
    print(f"{column}: {count}")
```

# Validate Emails Using Regular Expressions (re)

```python
import csv
import re

# Email pattern
pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

valid_emails = 0
invalid_emails = 0

with open("Orders_Cleaned.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        email = row["Email"]

        if re.match(pattern, email):
            valid_emails += 1
        else:
            invalid_emails += 1

print("Valid Emails :", valid_emails)
print("Invalid Emails :", invalid_emails)
```

# Calculate Revenue by Category

```python
import csv
# Dictionary to store category revenue
category_revenue = {}

with open("Orders_Cleaned.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        category = row["Category"]
        revenue = float(row["FinalAmount"])

        if category in category_revenue:
            category_revenue[category] += revenue
        else:
            category_revenue[category] = revenue

print("Revenue by Category:\n")

for category, revenue in sorted(category_revenue.items(), key=lambda x: x[1], reverse=True):
    print(f"{category}: {revenue:.2f}")
```

# Find Top 5 Most Frequently Purchased Products

```python
import csv
# Dictionary to count products
product_count = {}

with open("Orders_Cleaned.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        product = row["Product"]

        if product in product_count:
            product_count[product] += 1
        else:
            product_count[product] = 1

print("Top 5 Most Purchased Products:\n")

top_products = sorted(product_count.items(), key=lambda x: x[1], reverse=True)

for product, count in top_products[:5]:
    print(f"{product}: {count} orders")
```

# Generate Summary Report

```python
import csv

total_orders = 0
total_revenue = 0
total_rating = 0
customer_ids = set()

with open("Orders_Cleaned.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        total_orders += 1
        customer_ids.add(row["CustomerID"])
        total_revenue += float(row["FinalAmount"])
        total_rating += int(row["Rating"])

average_rating = total_rating / total_orders

print("========== SUMMARY REPORT ==========")
print("Total Orders        :", total_orders)
print("Unique Customers    :", len(customer_ids))
print("Total Revenue       :", round(total_revenue, 2))
print("Average Rating      :", round(average_rating, 2))
```
