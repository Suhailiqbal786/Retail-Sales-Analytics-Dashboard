# Handle Missing Values and Duplicates

```python
# Check missing values
print(orders.isnull().sum())
```

```python
# Check duplicate rows
print("Duplicate Rows:", orders.duplicated().sum())
```

```python
# Remove duplicates (if any)
orders = orders.drop_duplicates()
```

# Clean and Transform Columns

```python
# Convert Category to Proper Case
orders["Category"] = orders["Category"].str.title()

```

```python
# Trim CustomerName
orders["CustomerName"] = orders["CustomerName"].str.strip()

```

```python
# Convert OrderDate to datetime
orders["OrderDate"] = pd.to_datetime(orders["OrderDate"])
```

# Fix City Names and Ratings

```python
# Fix city names
orders["City"] = orders["City"].replace({
    "Hydrabad": "Hyderabad",
    "Banglaore": "Bangalore",
    "Delhl": "Delhi"
})
```

```python
orders["City"] = orders["City"].replace({
    "Chennnai": "Chennai"
})
```

```python
# Check unique city names
print("Unique Cities:")
print(sorted(orders["City"].unique()))

```

```python
# Check rating values
print("\nUnique Ratings:")
print(sorted(orders["Rating"].unique()))
```

# Perform Grouping and Analysis

```python
#1. Revenue by Category
category_revenue = orders.groupby("Category")["FinalAmount"].sum().sort_values(ascending=False)
print("Revenue by Category:")
print(category_revenue)
```

```python
#2. Top 5 Cities by Revenue
top_cities = orders.groupby("City")["FinalAmount"].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Cities:")
print(top_cities)
```

```python
# Top 5 Customers
top_customers = orders.groupby("CustomerName")["FinalAmount"].sum().sort_values(ascending=False).head(5)

print("Top 5 Customers:")
print(top_customers)
```

# Generate Summary Outputs

```python
print("========== SUMMARY ==========")

print("Total Orders:", len(orders))
print("Unique Customers:", orders["CustomerID"].nunique())
print("Total Revenue:", round(orders["FinalAmount"].sum(), 2))
print("Average Rating:", round(orders["Rating"].mean(), 2))

print("\nRevenue by Category:")
print(category_revenue)

print("\nTop 5 Cities:")
print(top_cities)

print("\nTop 5 Customers:")
print(top_customers)
```
